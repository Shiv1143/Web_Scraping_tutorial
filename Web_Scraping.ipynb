```python
import pandas as pd
```

```python
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Human_body_weight'
```

```python
response = requests.get(url, headers = {'user-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'})
```

```python
response
```

```python
response.text
```

```python
from bs4 import BeautifulSoup
```

```python
out = BeautifulSoup(response.text, "html.parser")
out
```

```python
out.find('table', {'class':'wikitable'})
```

```python
out.find_all('table', {'class':'wikitable'})
```

```python
len(out)
```

```python
l = out.find_all('table', {'class':'wikitable'})
```

```python
len(l)
```

```python
l[-1]
```

```python
data = pd.read_html(str(l[0]))
```

```python
data[0]
```

```python
df = data[0]
```

```python
df
```

```python
df.info()
```

```python
df.drop(columns = 'Source', inplace = True)
```

```python
df.head()
```

```python
df['Average weight']
```

```python
import re
```

```python
re.findall("\d+\. \d+", )
```

```python
df['Average weight'] = df['Average weight'].apply(lambda x: float(re.findall("\d+\.\d+", x.split("(")[0])[0]))
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns = {'Overweight population /  total population':'overweight(%age)'}, inplace = True)
```

```python
df.head()
```

```python
df['overweight(%age)'] = df['overweight(%age)'].apply(lambda x: float(re.findall("\d+\.\d+", x)[0]))
```
